```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.datasets import make_classification
import pandas as pd
import sqlite3
from contextlib import closing
```

```python
db = "test 2.db"
```

```python
def get_session():
    with closing(sqlite3.connect(db)) as conn:
        return pd.read_sql(
            f"SELECT * FROM datos", conn
        )
```

```python
session = get_session()
session
```

```python
size = session.index.size
entrenar = round((session.index.size)*0.7)
predecir = size - entrenar
print(size)
print(entrenar)
print(predecir)
```

```python
X = session[['delta0','theta0','low_alpha0','high_alpha0','low_beta0','high_beta0','low_gamma0','high_gamma0']].iloc[:entrenar,:]
y = session[['attention0','meditation0']].iloc[:entrenar,:]
```

```python
X
```

```python
y.head()
```

```python
clf = RandomForestClassifier(n_estimators=100, criterion='gini', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='sqrt', max_leaf_nodes=None, min_impurity_decrease=0.0, bootstrap=True, oob_score=False, n_jobs=None, random_state=None, verbose=0, warm_start=False, class_weight=None, ccp_alpha=0.0, max_samples=None, monotonic_cst=None)
clf.fit(X, y)
```

```python
Xp = session[['delta0','theta0','low_alpha0','high_alpha0','low_beta0','high_beta0','low_gamma0','high_gamma0']].iloc[entrenar:,:]
sizep = Xp.index.size
Xp
```

```python
predictions = pd.DataFrame(data=clf.predict(Xp), columns=["attention", "meditation"])
with closing(sqlite3.connect(db)) as conn:
    predictions.to_sql(name='randomforest', con=conn , if_exists='append')
```

```python
predictions
```

```python
comp = session[['attention0','meditation0']].iloc[entrenar:,:].rename(columns={"attention0": "attention", "meditation0": "meditation"})
comp.reset_index(drop=True, inplace=True)
```

```python
predictions.sub(comp, fill_value=0)
```
